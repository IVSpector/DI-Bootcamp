```python
# Exercise 1 : Use the openpyxl module to create a new workbook and add values/formulae to it

from openpyxl import Workbook

wb = Workbook()

ws = wb.active

ws['A1'] = 'First number ==>'
ws['B1'] = 2

ws['A2'] = 'Second number ==>'
ws['B2'] = 8

ws['A3'] = 'Total'
ws['B3'] = '=B1*B2'

wb.save("calculator.xlsx")

wb.close()
```

```python
# Exercise 2 : Open an Excel workbook and read down a list of plants finding the ones not in stock
from openpyxl import load_workbook
wb = load_workbook(filename='/content/Plants.xlsx')
ws = wb["Sheet1"]
this_plant = ws.cell(1,1)

while this_plant.offset(1,0).value:
    this_plant = this_plant.offset(1,0)
    stock_cell = this_plant.offset(0,7)
    if stock_cell.value == "No":
      print(this_plant.value)

wb.close()
```

```python
# Exercise 3 : Load Excel File and Write the Data
import pandas as pd
from openpyxl import load_workbook
from openpyxl.utils.dataframe import dataframe_to_rows

df = pd.read_excel('/content/data.xlsx')

df_filtered = df[df['Sales'] > 1000]

wb = load_workbook('/content/data.xlsx')

if 'Filtered' in wb.sheetnames:
    ws = wb['Filtered']
    wb.remove(ws)
ws = wb.create_sheet("Filtered")

for r_idx, row in enumerate(dataframe_to_rows(df_filtered, index=False, header=True)):
    ws.append(row)

wb.save('/content/data.xlsx')

wb.close()
```
