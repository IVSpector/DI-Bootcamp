```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('/content/database.sqlite')
cursor = conn.cursor()
# Получаем список всех таблиц
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
# Проходим по каждой таблице и выводим первые 5 строк
for table in tables:
    table_name = table[0]
    print(f'\n--- {table_name} ---')
    df = pd.read_sql_query(f"SELECT * FROM {table_name} LIMIT 5", conn)
    display(df)
```

```python
def sqlq(query):
    return pd.read_sql_query(query, conn)
```

```python
def print_all_table_columns():
# Prints the names of all tables and their columns
    cursor = conn.cursor()
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = [row[0] for row in cursor.fetchall()]
    for table in tables:
        print(f"--- {table} ---")
        cursor.execute(f"PRAGMA table_info('{table}')")
        columns = [col[1] for col in cursor.fetchall()]
        print(", ".join(columns))
        print()
    cursor.close()
print_all_table_columns()
```

```python
sql = '''
SELECT
  bb.Striker AS Batsman_Id,
  p.Player_Name,
  SUM(bs.Runs_Scored) AS Total_Runs
FROM Ball_by_Ball bb
JOIN Batsman_Scored bs
  ON bb.Match_Id = bs.Match_Id
  AND bb.Over_Id = bs.Over_Id
  AND bb.Ball_Id = bs.Ball_Id
JOIN Player p
  ON bb.Striker = p.Player_Id
GROUP BY bb.Striker, p.Player_Name
ORDER BY Total_Runs DESC;
'''
sqlq(sql)
```

```python
sql = '''
WITH innings_scores AS (
  SELECT
    bb.Striker AS Batsman_Id,
    p.Player_Name,
    bb.Match_Id,
    bb.Innings_No,
    SUM(bs.Runs_Scored) AS Runs_In_Innings
  FROM Ball_by_Ball bb
  JOIN Batsman_Scored bs
    ON bb.Match_Id = bs.Match_Id
    AND bb.Over_Id = bs.Over_Id
    AND bb.Ball_Id = bs.Ball_Id
  JOIN Player p
    ON bb.Striker = p.Player_Id
  GROUP BY bb.Striker, p.Player_Name, bb.Match_Id, bb.Innings_No
)

SELECT
  Batsman_Id,
  Player_Name,
  COUNT(CASE WHEN Runs_In_Innings BETWEEN 50 AND 99 THEN 1 END) AS Fifties,
  COUNT(CASE WHEN Runs_In_Innings >= 100 THEN 1 END) AS Hundreds
FROM innings_scores
GROUP BY Batsman_Id, Player_Name
ORDER BY Hundreds DESC, Fifties DESC;
'''
sqlq(sql)
```

```python
sql = '''
WITH bowler_wickets AS (
  SELECT
    bb.Bowler AS Bowler_Id,
    p.Player_Name,
    bb.Match_Id,
    wt.Innings_No,
    COUNT(*) AS Wickets
  FROM Wicket_Taken wt
  JOIN Ball_by_Ball bb
    ON wt.Match_Id = bb.Match_Id
    AND wt.Over_Id = bb.Over_Id
    AND wt.Ball_Id = bb.Ball_Id
  JOIN Player p
    ON bb.Bowler = p.Player_Id
  WHERE wt.Kind_Out IS NOT NULL AND wt.Player_Out IS NOT NULL
  GROUP BY bb.Bowler, p.Player_Name, bb.Match_Id, wt.Innings_No
),

best_figures AS (
  SELECT *,
    RANK() OVER (PARTITION BY Bowler_Id ORDER BY Wickets DESC) AS rnk
  FROM bowler_wickets
)

SELECT
  Bowler_Id,
  Player_Name,
  Match_Id,
  Innings_No,
  Wickets AS Best_Figures
FROM best_figures
WHERE rnk = 1
ORDER BY Best_Figures DESC;
'''
sqlq(sql)
```

```python
sql = '''
-- CTE 1: Total Runs
WITH total_runs AS (
  SELECT
    bb.Striker AS Player_Id,
    SUM(bs.Runs_Scored) AS Runs
  FROM Ball_by_Ball bb
  JOIN Batsman_Scored bs
    ON bb.Match_Id = bs.Match_Id AND bb.Over_Id = bs.Over_Id AND bb.Ball_Id = bs.Ball_Id
  GROUP BY bb.Striker
),

-- CTE 2: 50s and 100s
fifties_hundreds AS (
  SELECT
    bb.Striker AS Player_Id,
    COUNT(CASE WHEN SUM(bs.Runs_Scored) BETWEEN 50 AND 99 THEN 1 END) AS Fifties,
    COUNT(CASE WHEN SUM(bs.Runs_Scored) >= 100 THEN 1 END) AS Hundreds
  FROM Ball_by_Ball bb
  JOIN Batsman_Scored bs
    ON bb.Match_Id = bs.Match_Id AND bb.Over_Id = bs.Over_Id AND bb.Ball_Id = bs.Ball_Id
  GROUP BY bb.Striker, bb.Match_Id, bb.Innings_No
),

-- CTE 3: Total Wickets
total_wickets AS (
  SELECT
    bb.Bowler AS Player_Id,
    COUNT(*) AS Wickets
  FROM Wicket_Taken wt
  JOIN Ball_by_Ball bb
    ON wt.Match_Id = bb.Match_Id AND wt.Over_Id = bb.Over_Id AND wt.Ball_Id = bb.Ball_Id
  WHERE wt.Player_Out IS NOT NULL
  GROUP BY bb.Bowler
),

-- CTE 4: Best Bowling Figures
best_bowling_figures AS (
  SELECT Player_Id, Match_Id, Innings_No, Wickets
  FROM (
    SELECT
      bb.Bowler AS Player_Id,
      bb.Match_Id,
      wt.Innings_No,
      COUNT(*) AS Wickets,
      RANK() OVER (PARTITION BY bb.Bowler ORDER BY COUNT(*) DESC) AS rnk
    FROM Wicket_Taken wt
    JOIN Ball_by_Ball bb
      ON wt.Match_Id = bb.Match_Id AND wt.Over_Id = bb.Over_Id AND wt.Ball_Id = bb.Ball_Id
    WHERE wt.Player_Out IS NOT NULL
    GROUP BY bb.Bowler, bb.Match_Id, wt.Innings_No
  ) sub
  WHERE rnk = 1
),

-- CTE 5: Consolidated Player Metrics
career_metrics AS (
  SELECT
    p.Player_Id,
    p.Player_Name,
    COALESCE(r.Runs, 0) AS Total_Runs,
    COALESCE(f.Fifties, 0) AS Fifties,
    COALESCE(f.Hundreds, 0) AS Hundreds,
    COALESCE(w.Wickets, 0) AS Total_Wickets,
    COALESCE(bbf.Wickets, 0) AS Best_Bowling_Figures
  FROM Player p
  LEFT JOIN total_runs r ON p.Player_Id = r.Player_Id
  LEFT JOIN (
    SELECT Player_Id, COUNT(*) AS Fifties, COUNT(*) FILTER (WHERE Hundreds > 0) AS Hundreds
    FROM (
      SELECT Player_Id,
             CASE WHEN Runs BETWEEN 50 AND 99 THEN 1 END AS Fifties,
             CASE WHEN Runs >= 100 THEN 1 END AS Hundreds
      FROM (
        SELECT
          bb.Striker AS Player_Id,
          SUM(bs.Runs_Scored) AS Runs
        FROM Ball_by_Ball bb
        JOIN Batsman_Scored bs
          ON bb.Match_Id = bs.Match_Id AND bb.Over_Id = bs.Over_Id AND bb.Ball_Id = bs.Ball_Id
        GROUP BY bb.Striker, bb.Match_Id, bb.Innings_No
      ) x
    ) y
    GROUP BY Player_Id
  ) f ON p.Player_Id = f.Player_Id
  LEFT JOIN total_wickets w ON p.Player_Id = w.Player_Id
  LEFT JOIN best_bowling_figures bbf ON p.Player_Id = bbf.Player_Id
)

-- Final Select
SELECT *
FROM career_metrics
ORDER BY Total_Runs DESC, Total_Wickets DESC;
'''
sqlq(sql)
```
