```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://github.com/topics'

# Fetch the HTML content
response = requests.get(url)

# status_code
if response.status_code != 200:
    print(f"Failed to retrieve data. HTTP Status Code: {response.status_code}")
else:
    print(f"Status Code: {response.status_code}")

    # First 100 characters of the HTML content
    print(f"First 100 characters: {response.text[:100]}\n")

    # Saving into a file
    with open('webpage.html', 'w', encoding='utf-8') as f:
        f.write(response.text)

    # Parse the saved HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract titles of the topics and their descriptions (+ length)
    # Different 3 topics each time the web page is loaded
    topic_dict = {
        'title': [],
        'description': [],
        'length': [] # title + description
    }

    title = soup.find_all('p', class_='f3 lh-condensed text-center Link--primary mb-0 mt-1')
    description = soup.find_all('p', class_='f5 color-fg-muted text-center mb-0 mt-1')
    for t, d in zip(title, description):
        print(f"Title ({len(t.text.strip())}): {t.text.strip()}")
        print(f"Description ({len(d.text.strip())}): {d.text.strip()}")

         # Dictionary
        topic_dict['title'].append(t.text.strip())
        topic_dict['description'].append(d.text.strip())
        topic_dict['length'].append(len(t.text.strip()) + len(d.text.strip()))

    print(f"\n Dictionary :")
    for i in range(len(topic_dict['title'])):
        print(f" {topic_dict['title'][i]} ({topic_dict['length'][i]})")
        print(f"   {topic_dict['description'][i]}\n")

    # DataFrame
    df = pd.DataFrame(topic_dict)
```
